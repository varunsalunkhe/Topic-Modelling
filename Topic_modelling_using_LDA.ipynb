```python
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
data = pd.read_csv("quora_questions.csv")
data.head()
```

```python
data.shape
```

```python
data = data.iloc[:10000,:]
data.shape
```

```python
data.head()
```

```python
vect_coverter = CountVectorizer(max_df = 0.9, min_df = 1,stop_words= "english")
dmat = vect_coverter.fit_transform(data['Question'])
dmat
```

```python
lda = LatentDirichletAllocation(n_components = 10 , random_state = 32)
lda.fit(dmat)
```

```python
len(vect_coverter.get_feature_names_out())
```

```python
lda.components_
```

```python
lda.components_.shape
```

```python
#printing most relevant words of each topic

for i, words in enumerate(lda.components_):
  print(f"for #{i}")
  print([vect_coverter.get_feature_names_out()[i] for i in words.argsort()[-15:]])
  print("\n")
```

```python
final_trans= lda.fit_transform(dmat)
final_trans.shape
```

```python
final_trans[0]
```

```python
data["cluster"] = final_trans.argmax(axis=1)
data.head()
```
